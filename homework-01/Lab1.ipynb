```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import random as rd
```

```python
def f(xx):
    return (np.sin(xx / 5) * (np.exp(xx / 10)) + 5 * np.exp(-0.5 * xx))
```

## Полиномиальная регрессия

```python
xx = np.linspace(1, 15, 5)
yy = f(xx)
def comp_pol_coef(xx, yy):
    n = len(xx)
    b = yy
    A = np.zeros((n,n))
    for i in range(n):
        A[:, i] = xx ** i
    return np.linalg.solve(A, b)
def polynome(w, x):
    s = 0
    for i in range(len(w)):
        s += w[i] * (x ** i)
    return s
```

```python
n = 15
w = comp_pol_coef(xx,yy)
x_points = np.linspace(1, 15, n)
y_points = f(x_points)
xx = np.linspace(1, 15, 100)
plt.plot(xx, f(xx), label = 'f(x)')
xx = np.linspace(1, 15, 100)
plt.plot(xx, polynome(w, xx), '--', label = 'polynome')
plt.plot(x_points, y_points, 'o', label = 'points')
plt.legend()
plt.show()
```

## Полиномиальная регрессия с зашумлённым данными

```python
n = 15
noise = 0.01
x_points = np.linspace(1, 15, n)
y_points = f(x_points) + np.random.randn(n) * noise
w = comp_pol_coef(x_points, y_points)
xx = np.linspace(1, 15, 100)
plt.plot(xx, f(xx), label = 'f(x)')
xx = np.linspace(1, 15, 100)
plt.plot(xx, polynome(w, xx), '--', label = 'polynome')
plt.plot(x_points, y_points, 'o', label = 'points')
plt.legend()
plt.show()
```

## Метод наименьших квадратов

```python
def f(xx):
    return (np.sin(xx / 5) * (np.exp(xx / 10)) + 5 * np.exp(-0.5 * xx))
N = 100
xx = np.linspace(1, 15, N)
yy = f(xx)
def method_naim_kvadratov(xx, yy):
    n = len(xx)
    b = yy
    A = np.zeros((N,n))
    for i in range(n):
        A[:, i] = xx ** i
        t = np.linalg.lstsq(A, b)
    return t[0]
def polynome(w, x):
    s = 0
    for i in range(len(w)):
        s += w[i] * (x ** i)
    return s
```

```python
n = 15
noise2 = 0.1
w = method_naim_kvadratov(xx, yy)
x_points = np.linspace(1, 15, n)
y_points = f(x_points) + np.random.randn(n) * noise2
xx = np.linspace(1, 15, N)
plt.plot(xx, f(xx), label = 'f(x)')
xx = np.linspace(1, 15, N)
plt.plot(xx, polynome(w, xx), '--', label = 'polynome')
plt.plot(x_points, y_points, 'o', label = 'points')
plt.legend()
plt.show()
```
